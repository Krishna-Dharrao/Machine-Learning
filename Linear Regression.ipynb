```python
x = np.random.rand(50,1)
x
```

```python
y = 3*x+30+np.random.rand(50,1)
y
```

```python
#estimating coefficents
def coeff(x,y):
    cov = 0.0
    var = 0.0
    for i in range(50):
        cov =+(x[i]-np.mean(x))*(y[i]-np.mean(y))
        var =+pow((x[i]-np.mean(x)),2)
    print(cov,var)
    b1=cov/var
    b0=np.mean(y)-b1*np.mean(x)
    return b0,b1
b0,b1=coeff(x,y)
```

```python
#fitted linear equation / predicted y
y1 = b0+b1*x
y1
```

```python
#plotting the fitted line
plt.scatter(x,y, color='black')
plt.plot(x,y1, color='blue')
plt.show()
```
